```python
# Here are our two 3x3 matrices (lists of lists)
matA = [[1, 2, 3],
        [4, 5, 6],
        [7, 5, 6]]

matB = [[4, 6, 2],
        [2, 7, 6],
        [3, 4, 6]]

# To store our results, we need to create empty 3x3 matrices.
# Let's fill them with zeros to start.

# --- Create an empty result matrix for addition ---
addition_result = []
for i in range(3):
    # For each row, add a new empty list
    row = [0, 0, 0]
    addition_result.append(row)

# --- Create an empty result matrix for subtraction ---
subtraction_result = []
for i in range(3):
    row = [0, 0, 0]
    subtraction_result.append(row)

# --- Create an empty result matrix for multiplication ---
multiplication_result = []
for i in range(3):
    row = [0, 0, 0]
    multiplication_result.append(row)


# Now, let's loop through each cell of the matrices to perform the calculations.
# 'i' will be the row number (0, 1, or 2)
# 'j' will be the column number (0, 1, or 2)
for i in range(3):
    for j in range(3):
        # Add the elements at the same position
        addition_result[i][j] = matA[i][j] + matB[i][j]
        
        # Subtract the elements at the same position
        subtraction_result[i][j] = matA[i][j] - matB[i][j]
        
        # Multiply the elements at the same position
        multiplication_result[i][j] = matA[i][j] * matB[i][j]

# --- Finally, print everything out nicely ---
print("Matrix Addition (A + B):")
for row in addition_result:
    print(row)

print("\nMatrix Subtraction (A - B):")
for row in subtraction_result:
    print(row)

print("\nElement-wise Multiplication (A * B):")
for row in multiplication_result:
    print(row)
```

```python
# Here is our nested tuple
data = (
    (1, 5, 3),      # sum = 9
    (10, 20),       # sum = 30
    (7, 8, 9, 10),  # sum = 34
    (2, 4, 6, 8),   # sum = 20
    (11, 12, 8)     # sum = 31
)

# Step 1: Create an empty list to hold the sum of each inner tuple.
sums_list = []

# Loop through each tuple inside 'data'
for inner_tuple in data:
    # Calculate the sum of the current tuple
    current_sum = sum(inner_tuple)
    # Add this sum to our list
    sums_list.append(current_sum)

# Step 2: Sort the list of sums to easily find the largest and second largest.
# The sort() method sorts the list from smallest to largest.
sums_list.sort()

# The second largest sum will be the second to last item in the sorted list.
second_largest_sum = sums_list[-2]

# Step 3: Now we need to find which original tuple has this sum.
# We'll loop through the original 'data' again.
result_tuple = None # A variable to store our answer
for inner_tuple in data:
    # If the sum of this tuple matches our target sum...
    if sum(inner_tuple) == second_largest_sum:
        # ...we've found our answer!
        result_tuple = inner_tuple
        break # We can stop looping now

print(f"The nested tuples are: {data}")
print(f"The sorted list of sums is: {sums_list}")
print(f"The second largest sum is: {second_largest_sum}")
print(f"The tuple with that sum is: {result_tuple}")
```

```python
# Here is our dictionary of students and their marks
students = {
    'Alice': {'Math': 90, 'Science': 88, 'History': 92},
    'Bob': {'Math': 78, 'Science': 80},
    'Charlie': {'Math': 95, 'Science': 91, 'History': 85}
}

# We need two empty dictionaries:
# one to store the total score for each subject...
subject_totals = {}
# ...and one to store how many students took each subject.
subject_counts = {}

# Loop through each student in the main dictionary
for student_name, marks in students.items():
    # Now loop through the subjects and scores for that student
    for subject, score in marks.items():
        
        # Check if we have seen this subject before
        if subject in subject_totals:
            # If yes, add the current score to the existing total
            subject_totals[subject] = subject_totals[subject] + score
            subject_counts[subject] = subject_counts[subject] + 1
        else:
            # If no, this is the first time we're seeing this subject.
            # So, create it in our dictionaries.
            subject_totals[subject] = score
            subject_counts[subject] = 1

# Now, calculate the average for each subject
subject_averages = {}
for subject, total_score in subject_totals.items():
    count = subject_counts[subject]
    average = total_score / count
    subject_averages[subject] = average


print("\nAverage score for each subject:")
print(subject_averages)
```

```python
# 1. Let's create a dictionary. It stores things in key-value pairs.
# Think of it like a contact list: 'name' is a key, 'John' is its value.
my_dict = {'name': 'John', 'age': 25, 'city': 'New York'}
print(f"Our starting dictionary: {my_dict}")

# 2. To get a value, you use its key in square brackets.
persons_name = my_dict['name']
print(f"Getting the value for 'name': {persons_name}")

# 3. To add a new item, just define it.
my_dict['occupation'] = 'Engineer'
print(f"Added a new item 'occupation': {my_dict}")

# 4. To change a value, just assign a new one to the key.
my_dict['age'] = 26
print(f"Changed the value of 'age': {my_dict}")

# 5. To remove an item, you can use 'del'.
del my_dict['city']
print(f"Removed the 'city' item: {my_dict}")

# 6. You can check if a key exists in the dictionary. This gives True or False.
is_name_present = 'name' in my_dict
print(f"Is the key 'name' in our dictionary? {is_name_present}")

# 7. Get a list of all the keys.
all_keys = my_dict.keys()
print(f"All the keys are: {all_keys}")

# 8. Get a list of all the values.
all_values = my_dict.values()
print(f"All the values are: {all_values}")

# 9. Get the number of items in the dictionary.
num_items = len(my_dict)
print(f"Number of items: {num_items}")

# 10. You can wipe the dictionary clean.
my_dict.clear()
print(f"The dictionary after clearing it: {my_dict}")
```

```python
sentence = "The cat sat on the mat. The mat was by the door, and the cat looked at the door."

# Step 1: Clean the sentence.
# First, make everything lowercase so 'The' and 'the' are treated the same.
lower_sentence = sentence.lower()

# Second, remove punctuation. We can do this by replacing it with nothing.
clean_sentence = lower_sentence.replace('.', '')
clean_sentence = clean_sentence.replace(',', '') # Add more if needed

# Step 2: Split the sentence into a list of words.
words = clean_sentence.split()

# Step 3: Count each word's frequency.
# We'll use a dictionary where the word is the key and the count is the value.
word_counts = {}

# Loop through every word in our list of words
for word in words:
    # Check if we have already counted this word
    if word in word_counts:
        # If yes, just add 1 to its current count
        word_counts[word] = word_counts[word] + 1
    else:
        # If no, this is the first time we've seen this word. Add it to
        # the dictionary and set its count to 1.
        word_counts[word] = 1

# Step 4: Find the highest count.
max_count = 0
# Loop through the counts (the values) in our dictionary
for count in word_counts.values():
    # If we find a count that is bigger than our current max...
    if count > max_count:
        # ...it becomes the new max!
        max_count = count

# Step 5: Find all words that have this highest count.
# Create an empty list to store the winning words.
most_frequent_words = []
# Loop through the dictionary, getting both the word and its count
for word, count in word_counts.items():
    # If this word's count is the same as the max count...
    if count == max_count:
        # ...add it to our list of winners.
        most_frequent_words.append(word)

# --- Print all our results ---
print(f"Original Sentence: '{sentence}'")
print(f"\nCleaned list of words: {words}")
print(f"\nDictionary of word counts: {word_counts}")
print(f"\nThe highest count found was: {max_count}")
print(f"The most frequent word(s) are: {most_frequent_words}")
```
